```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from  keras.preprocessing import image
from keras.applications import resnet50
```

```python
model = resnet50.ResNet50()
```

### A random dog's pic 

```python
img = image.load_img("cat_or_dog_1.jpg",target_size=(224,224))
```

```python
img
```

```python
x = resnet50.preprocess_input(np.expand_dims(image.img_to_array(img),axis=0))
```

```python
predictions = model.predict(x)
classes = resnet50.decode_predictions(predictions)
```

```python
print(f"{list(classes[0][0])[1]} with accuracy {list(classes[0][0])[2]*100} %")
print(f"{list(classes[0][1])[1]} with accuracy {list(classes[0][1])[2]*100} %")
print(f"{list(classes[0][2])[1]} with accuracy {list(classes[0][2])[2]*100} %")
```
